### 1. Label & Button

```python
import tkinter as tk

# Tk is an object
window = tk.Tk()  
window.title('my window')
window.geometry('200x100')

# StringVar is a variable
var = tk.StringVar()
flg = 0

# Lable is also an object, with initial input window
lb = tk.Label(window,
              # text = 'I am a Label',
              textvariable = var,
              bg = 'green', 
              font = ('Arial', 12),
              width = 12, # The unit of width and hight is word
              height = 2)
lb.pack()  # Place the label

def hit_me():
    global flg
    if flg == 0:
        var.set('Hahaha')
        flg = 1
    else:
        var.set('')
        flg = 0

# Button is the third object, with input window
bt = tk.Button(window,
              text = 'I am a Botton', 
              width = 12,
              height = 2,
              command = hit_me)
bt.pack()  # Place the label

window.mainloop()  # Dynamic logit
```

### 2. Entry & Text

```python
import tkinter as tk

# Tk is an object
window = tk.Tk()  
window.title('my window')
window.geometry('200x200')

# Entry
et = tk.Entry(window, show = None)  # show = '*'
et.pack()

# Button 1
def insert_point():
    var = et.get()
    tx.insert('insert', var)
    
bt1 = tk.Button(window,
              text = 'Insert Point', 
              width = 12,
              height = 2,
              command = insert_point)
bt1.pack()  # Place the label

# Button 2
def insert_end():
    var = et.get()
    tx.insert('end', var)
    
bt2 = tk.Button(window,
              text = 'Insert End', 
              width = 12,
              height = 2,
              command = insert_end)
bt2.pack()  # Place the label

# Text
tx = tk.Text(window, height = 2)
tx.pack()

window.mainloop()  # Dynamic logit
```

### 3. ListBox

```python
import tkinter as tk

# Tk is an object
window = tk.Tk()  
window.title('my window')
window.geometry('200x220')

var1 = tk.StringVar()

# Label
lb = tk.Label(window,
              textvariable = var1,
              bg = 'yellow',
              font = ('Arial', 12),
              width = 12,
              height = 2)
lb.pack()

# ListBox
my_list = ('Zhouqiao', 'Jun', 'Fungart', 'Mooniu')
var2 = tk.StringVar()
var2.set(my_list)
listbox = tk.Listbox(window, listvariable = var2, height = 6)
list_new_items = ['Panda', 'Bob']
for item in list_new_items:
    listbox.insert('end', item)
listbox.insert(1, 'Xiaoming')
listbox.delete(1)
listbox.pack()

# Button
def print_sel():
    value = listbox.get(listbox.curselection())
    var1.set(value)
    
bt = tk.Button(window,
              text = 'Print Selection', 
              width = 12,
              height = 2,
              command = print_sel)
bt.pack()  # Place the label

window.mainloop()  # Dynamic logit
```

### 4. Radiobutton

```python
import tkinter as tk

# Tk is an object
window = tk.Tk()  
window.title('my window')
window.geometry('200x150')

var = tk.StringVar()

# Label
lb = tk.Label(window,
              text = '',
              bg = 'yellow',
              font = ('Arial', 12),
              width = 40,
              height = 2)
lb.pack()

# Radiobutton
def select_b():
    lb.config(text = 'You have selected ' + var.get())
rb1 = tk.Radiobutton(window,
                     text = 'Option A',
                     variable = var,
                     value = 'A',
                     command = select_b)
rb2 = tk.Radiobutton(window,
                     text = 'Option B',
                     variable = var,
                     value = 'B',
                     command = select_b)
rb3 = tk.Radiobutton(window,
                     text = 'Option C',
                     variable = var,
                     value = 'C',
                     command = select_b)
rb1.pack()
rb2.pack()
rb3.pack()
window.mainloop()  # Dynamic logit
```

### 5. Scale

```python
import tkinter as tk

# Tk is an object
window = tk.Tk()
window.title('my window')
window.geometry('200x100')

# Label
lb = tk.Label(window,
              bg = 'yellow',
              width = 20,
              text = '')
lb.pack()

def print_selection(value):
    lb.config(text='you have selected ' + value)

# Scale
sc = tk.Scale(window,
              label = 'Move Me',
              orient = tk.HORIZONTAL,
              from_ = 5,
              to = 11,
              length = 200,
              showvalue = 0,
              tickinterval = 2,
              resolution = 0.01,
              command = print_selection)
sc.pack()

window.mainloop()
```

### 6. Checkbutton

```python
import tkinter as tk

# window
window = tk.Tk()
window.title('My window')
window.geometry('200x100')

# Set variable
var_t = tk.StringVar()
var_t.set('Choose Coding Language')
var_i1 = tk.IntVar()
var_i2 = tk.IntVar()

# Label
lb = tk.Label(window,
              bg = 'yellow',
              width = 20,
              height = 2,
              textvariable = var_t)
lb.pack()

# Checkbutton
def print_str():
    if var_i1.get()==0 and var_i2.get()==0:
        var_t.set('I do not love eigher')
    elif var_i1.get()==1 and var_i2.get()==0:
        var_t.set('I love C++')
    elif var_i1.get()==0 and var_i2.get()==1:
        var_t.set('I love Python')
    elif var_i1.get()==1 and var_i2.get()==1:
        var_t.set('I love both')
cb1 = tk.Checkbutton(window,
                     text = 'C++',
                     variable = var_i1,
                     onvalue = 1, offvalue = 0,
                     command = print_str)
cb2 = tk.Checkbutton(window,
                     text = 'Python',
                     variable = var_i2,
                     onvalue = 1, offvalue = 0,
                     command = print_str)
cb1.pack()
cb2.pack()

window.mainloop()
```

### 7. Canvas

```python
import tkinter as tk

# Window
window = tk.Tk()
window.title('my window')
window.geometry('200x200')

# Canvas
canvas = tk.Canvas(window,
                   bg='blue',
                   height=100,
                   width=200)
image_file = tk.PhotoImage(file = 'bridge.png')
image = canvas.create_image(10, 10, anchor = 'nw', image = image_file)
x0, y0, x1, y1= 50, 50, 80, 80
line = canvas.create_line(x0, y0, x1, y1)
oval = canvas.create_oval(x0, y0, x1, y1, fill='red')
arc = canvas.create_arc(x0+30, y0+30, x1+30, y1+30, start=0, extent=180)
rect = canvas.create_rectangle(100, 30, 100+20, 30+20)
canvas.pack()

def moveit():
    canvas.move(rect, 0, 2)

# Button
b = tk.Button(window,
              text='move',
              command=moveit)
b.pack()


window.mainloop()
```

### 8. Menubar

```python
import tkinter as tk

# Window
window = tk.Tk()
window.title('my window')
window.geometry('200x200')

# Label
l = tk.Label(window, text='', bg='yellow')
l.pack()
counter = 0

# Show
def do_job():
    global counter
    l.config(text='do '+ str(counter))
    counter+=1

# Menubar
menubar = tk.Menu(window)
filemenu = tk.Menu(menubar, tearoff=0)
menubar.add_cascade(label='File', menu=filemenu)
filemenu.add_command(label='New', command=do_job)
filemenu.add_command(label='Open', command=do_job)
filemenu.add_command(label='Save', command=do_job)
filemenu.add_separator()
filemenu.add_command(label='Exit', command=window.quit)

editmenu = tk.Menu(menubar, tearoff=0)
menubar.add_cascade(label='Edit', menu=editmenu)
editmenu.add_command(label='Cut', command=do_job)
editmenu.add_command(label='Copy', command=do_job)
editmenu.add_command(label='Paste', command=do_job)

submenu = tk.Menu(filemenu)
filemenu.add_cascade(label='Import', menu=submenu, underline=0)
submenu.add_command(label="Submenu1", command=do_job)

window.config(menu=menubar)

window.mainloop()
```

### 9. Frame

```python
import tkinter as tk

# Window
window = tk.Tk()
window.title('my window')
window.geometry('200x100')

# Label
tk.Label(window, text='on the window').pack()

# Frame
frm = tk.Frame(window)
frm.pack()

frm_l = tk.Frame(frm, )
frm_r = tk.Frame(frm)
frm_l.pack(side='left')
frm_r.pack(side='right')

tk.Label(frm_l, text='on the frm_l1').pack()
tk.Label(frm_l, text='on the frm_l2').pack()
tk.Label(frm_r, text='on the frm_r1').pack()

# Run
window.mainloop()
```

### 10. Messagebox

```python
import tkinter as tk
import tkinter.messagebox

window = tk.Tk()
window.title('my window')
window.geometry('200x200')

def hit_me():
    print(tk.messagebox.showinfo(title='Hi', message='hahahaha'))   # return 'ok'
    print(tk.messagebox.showwarning(title='Hi', message='nononono'))   # return 'ok'
    print(tk.messagebox.showerror(title='Hi', message='No!! never'))   # return 'ok'
    print(tk.messagebox.askquestion(title='Hi', message='hahahaha'))   # return 'yes' , 'no'
    print(tk.messagebox.askyesno(title='Hi', message='hahahaha'))   # return True, False
    print(tk.messagebox.askretrycancel(title='Hi', message='hahahaha'))   # return True, False
    print(tk.messagebox.askokcancel(title='Hi', message='hahahaha'))   # return True, False
    print(tk.messagebox.askyesnocancel(title="Hi", message="haha"))     # return, True, False, None

tk.Button(window, text='hit me', command=hit_me).pack()
window.mainloop()
```
