# Dict 的一些操作

```python
Dict={0:"Car", 1:"Bicycle", 2:"Boat"}
```

```python
len(Dict)
```

```python
if 0 in Dict:#检验
    print(Dict.get(0))#打印
```

```python
Dict[0]#Dict.get()
```

```python
Dict['a'] = "Plane"
```

```python
len(Dict)
```

```python
print(Dict["a"])
#这是一个多行注释：
#Dictionary 即map结构
#这是一种无须结构
#key 可以是字符串
#python 没有 char类型
#单双引号可以混用
```

```python
for k in Dict:
    print(Dict.get(k))
```

```python
if 0 in Dict:
    del Dict[0]    #有检验的delete
for k in Dict:
    print(k)    #key
    print(Dict.get(k))    #value
    
print(  )

if 1 in Dict:
    print(Dict.pop(1))
    
print(  )

for k in Dict:
    print(k)
    print(Dict.get(k))
```

```python
a = Dict.get(2,"傻逼")
b = Dict.get(0,"傻逼")
print(a+b)
```

```python
Dict.items()
#注意返回格式list（tuple）
for (x,y) in Dict.items():#()可省
    print("\n", x)
    print(y)
```
